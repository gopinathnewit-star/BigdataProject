```python
import requests
import pandas as pd
import json

# ------------------------------------------
# GNEWS API
# ------------------------------------------
API_KEY = "api_from_env"
query = "technology"
url = f"https://gnews.io/api/v4/search?q={query}&token={API_KEY}"

response = requests.get(url)
data = response.json()

# ------------------------------------------
# Save RAW Bronze File
# ------------------------------------------
bronze_path = "/Workspace/social media/Bronze/news_raw.json"

with open(bronze_path, "w") as f:
    json.dump(data, f, indent=4)

print("Bronze layer raw file saved:", bronze_path)
```
